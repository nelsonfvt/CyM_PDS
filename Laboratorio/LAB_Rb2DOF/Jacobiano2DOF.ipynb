Jacobiano del robot 2DOF

El jacobiano del robot permite mapear el espacio de velocidades de las articulaciones hacia el espacio de velocidades del efector final (punta, pinza).

A continuación se defina la matriz Jacobiana y se desarrolla un ejemplo muy sencillo de aplicación


```python
import numpy as np
from D_Jacobian import D_Jacobian
```

```python
dt = 0.01 # periodo de muestreo
q = np.array([np.pi/3, np.pi/3]) #ángulos iniciales

#longitudes eslabones
l_1 = 10
l_2 = 10

ft = 0.1

t = np.arange(0, 10.01, dt)

v_p = np.zeros((2, len(t)))
q_p = np.zeros((2, len(t)))

for k in range(1,len(t)):
    q_p[:,k-1] = [ft - ft * np.cos(2*np.pi*2*t[k]), ft - ft * np.cos(2*np.pi*1*t[k])]

    J = D_Jacobian(q, np.array([l_1, l_2]))

    v_p[:,k-1] = J.dot(q_p[:,k-1])

    q = q + q_p[:,k-1] * dt
print(q_p)
```

Jacobiano inverso

Mediante la inversión de la matriz jacobiana es posible mapear desde el espacio de velocidades del efector final hacia el espacio de velocidades de las articulaciones.

```python
q = np.array([np.pi/3, np.pi/3]) #ángulos iniciales
q_p2 = np.zeros((2, len(t)))

for k in range(1,len(t)):
    J_a = D_Jacobian(q, np.array([l_1, l_2]))
    J_i = np.linalg.inv(J_a)

    q_p2[:, k] = J_i.dot(v_p[:, k-1])

    q = q + q_p2[:, k-1] * dt
print(q_p2)
```
