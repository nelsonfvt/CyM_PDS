```python
import numpy as np
import matplotlib.pyplot as plt
import cmath
```

Combinación de varias funciones exponenciales complejas para generar otra señal

```python
n = np.arange(0,15)
A1 = 3.0
p1 = 0.2
A2 = -0.2
p2 = -0.1
A3 = 0.1
f1 = np.pi/3.0 # valores entre -pi y pi
xnv = A1*(p1**n) + A2*(p2**n) + A3*complex(np.cos(f1),np.sin(f1))**n + A3*complex(np.cos(f1),-np.sin(f1))**n
plt.figure(1)
plt.stem(n, xnv.real) 
plt.title("Señal resultante") 
plt.xlabel("muestras") 
plt.ylabel("Amplitud") 
plt.show()
```

Ubicación de los números complejos en el plano

```python
def circulo(x,y,r):
    ang = np.arange(0,2*np.pi,0.01)
    xp = r*np.cos(ang)
    yp = r*np.sin(ang)
    plt.plot(x+xp,y+yp,color = 'red')
```

```python
z1 = complex(np.cos(f1),np.sin(f1))
z2 = complex(np.cos(f1),-np.sin(f1))
plt.figure(2)
circulo(0,0,1)
plt.scatter(p1,0,marker='x',color='blue')
plt.scatter(p2,0,marker='x',color='blue')
plt.scatter(z1.real,z1.imag,marker='x',color='blue')
plt.scatter(z2.real,z2.imag,marker='x',color='blue')
plt.grid()
plt.xlim(-2, 2)
plt.ylim(-2, 2)
plt.xlabel("Real") 
plt.ylabel("Imaginario") 
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```
