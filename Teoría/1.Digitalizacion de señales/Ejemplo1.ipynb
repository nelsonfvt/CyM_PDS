```python
import numpy as np
import matplotlib.pyplot as plt
```

Generando señal de tiempo continuo.

Se tienen dos componentes:

4 Hz

6 Hz

```python
t=np.arange(0.0,1.0,0.001)
f1 = 4 #Hz
f2 = 6 #Hz
xt = 2*np.cos(2*np.pi*f1*t) + 3*np.sin(2*np.pi*f2*t)
plt.figure(1)
plt.plot(t, xt, color = 'blue') 
plt.title("Señal análoga") 
plt.xlabel("Tiempo (s)") 
plt.ylabel("Amplitud") 
plt.show()
```

Discretización de la señal análoga.

Se define una frecuencia de muestreo de 24 muestras/segundo

```python
Fs = 24.0
Ts=1.0/Fs
nT=np.arange(0.0,1.0,Ts)
xnT= 2*np.cos(2*np.pi*f1*nT) + 3*np.sin(2*np.pi*f2*nT)
```

Mostrando grafica de senal continua y de los instantes de muestreo

```python
plt.figure(2)
plt.plot(t, xt, color = 'red') 
plt.stem(nT,xnT) 
plt.title("Discretización de señal análoga") 
plt.xlabel("Tiempo (s)") 
plt.ylabel("Amplitud") 
plt.show()
```

Se genera la senal de tiempo discreto correspondiente:

Observe que no depende del tiempo pero si de las muestras

```python
N=Fs
n=np.arange(0, N,1.0)
xn = 2*np.cos(2*np.pi*(f1/Fs)*n) + 3*np.sin(2*np.pi*(f2/Fs)*n)
```

Generando grafico de la senal de tiempo discreto

```python
plt.figure(3)
plt.stem(n, xn)
plt.title("Señal de tiempo discreto")
plt.xlabel("Muestras") 
plt.ylabel("Amplitud") 
plt.xticks(n,range(len(n)))
plt.show()
```
