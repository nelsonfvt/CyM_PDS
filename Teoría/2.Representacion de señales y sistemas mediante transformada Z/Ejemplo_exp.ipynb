```python
import numpy as np
import matplotlib.pyplot as plt
import cmath
```

Señal exponencial real

Dependiendo del valor de la base (A) esta función tendrá una forma diferente.
Pruebe diferentes valores de A para observar lo que ocurre

```python
n = np.arange(0,10)
A = 0.5
zxn = A**n
plt.figure(1)
plt.stem(n, zxn) 
plt.title("Señal exponencial") 
plt.xlabel("muestras") 
plt.ylabel("Amplitud") 
plt.show()
```

Ahora se consdera que la base de la función exponencial es un número complejo

```python
zreal = -0.6
zimag = -0.4
z = complex(zreal,zimag)
zxn = z**n
```

Graficando la parte real

```python
plt.figure(2)
plt.stem(n, zxn.real) 
plt.title("Señal exponencial parte real") 
plt.xlabel("muestras") 
plt.ylabel("Amplitud") 
plt.show()
```

Graficando la parte imaginaria

```python
plt.figure(3)
plt.stem(n, zxn.imag) 
plt.title("Señal exponencial parte imaginaria") 
plt.xlabel("muestras") 
plt.ylabel("Amplitud") 
plt.show()
```

Localización de z en el plano complejo

```python
def circulo(x,y,r):
    ang = np.arange(0,2*np.pi,0.01)
    xp = r*np.cos(ang)
    yp = r*np.sin(ang)
    plt.plot(x+xp,y+yp,color = 'red')
```
