```python
import numpy as np
import cmath
import matplotlib.pyplot as plt
```

Ejemplo de aplicación de la FFT a una señal periodica.

En este ejemplo se toma como base la misma señal de ejemplo del archivo formulado en excel. 

```python
N = 128
n = np.arange(0,N)

# Armonicos
Ar1 = 1.0 * np.cos(2*np.pi* 1.0/4.0 *n)
Ar2 = 1.0 * np.sin(2*np.pi* 1.0/8.0 *n)
Ar3 = 0.0 * np.cos(2*np.pi* 1.0/16.0 *n)
Ar4 = 0.0 * np.sin(2*np.pi* 1.0/32.0 *n)
Ar5 = 0.0 * np.cos(2*np.pi* 1.0/64.0 *n)
# Señal a partir de armónicos
xn = Ar1+Ar2+Ar3+Ar4+Ar5

#gráfica
plt.figure(1)
plt.stem(n,xn)
plt.xlabel('muestras')
plt.ylabel('Amplitud')
plt.show()
```

Aplicado la fft a la señal generada

```python
Xk = np.fft.fft(xn)

# Parte real e imaginaria
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True)
ax1.set_title('Parte real')
ax1.stem(n,Xk.real,linefmt='red')
ax1.set_xlabel('muestras')
ax2.set_title('Parte imaginaria')
ax2.stem(n,Xk.imag,linefmt='blue')
ax2.set_xlabel('muestras')

# Magnitud y fase
f, (ax3, ax4) = plt.subplots(1, 2, sharey=True)
ax3.set_title('Magnitud')
ax3.stem(n,abs(Xk),linefmt='red')
ax3.set_xlabel('muestras')
ax4.set_title('Fase')
ax4.stem(n,np.angle(Xk),linefmt='blue')
ax4.set_xlabel('muestras')
plt.show()
```

Para obtener los mismos resultados que lo observado en el archivo de excel,
hay que dividir los valores obtenidos con la fft entre el número total de muestras

```python
XkN=Xk/N

# Parte real e imaginaria
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True)
ax1.set_title('Parte real')
ax1.stem(n,XkN.real,linefmt='red')
ax1.set_xlabel('muestras')
ax2.set_title('Parte imaginaria')
ax2.stem(n,XkN.imag,linefmt='blue')
ax2.set_xlabel('muestras')

# Magnitud y fase
f, (ax3, ax4) = plt.subplots(1, 2, sharey=True)
ax3.set_title('Magnitud')
ax3.stem(n,abs(XkN),linefmt='red')
ax3.set_xlabel('muestras')
ax4.set_title('Fase')
ax4.stem(n,np.angle(XkN),linefmt='blue')
ax4.set_xlabel('muestras')
plt.show()
```
