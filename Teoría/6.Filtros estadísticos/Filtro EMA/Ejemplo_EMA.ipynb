```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
```

Ejemplo de diseño y aplicación de filtro Media movil exponencial (EMA)

Inicialmente se crea una señal de prueba con varios armónicos para observar el desempeño de los filtros
También se calcula su espectro de frecuencia usando la FFT

```python
n = np.arange(0,512)
k = n/512
xn=np.sin(2*np.pi*n*1/4)+np.sin(2*np.pi*n*1/8)+np.sin(2*np.pi*n*1/32)+np.sin(2*np.pi*n*1/128)
Xk = abs(np.fft.fft(xn))

# Graficos

#señal
plt.figure(1)
plt.stem(n,xn)
plt.xlabel('muestras')
plt.ylabel('Amplitud')

#magnitud del espectro
plt.figure(2)
plt.stem(k,Xk)
plt.xlabel('Frec. Normalizada')
plt.ylabel('Amplitud armónicos')
plt.show()
```

```python
# generando el filtro
alpha = 0.1
b = alpha
a = np.array([1, -(1-alpha)])

# Filtrar la señal
yn = signal.lfilter(b,a,xn)
plt.figure()
plt.stem(n,yn)
plt.xlabel('muestras')
plt.ylabel('Amplitud')
plt.title('Señal filtrada')

# Espectro de frecuencia
plt.figure()
plt.stem(k,abs(np.fft.fft(yn)))
plt.xlabel('Frec. Normalizada')
plt.ylabel('Amplitud armónicos')
plt.show()
```
